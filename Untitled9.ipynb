```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Bidirectional, Dense, Dropout

# 1. NLTK Resources Download
nltk.download('stopwords')
ps = PorterStemmer()
stop_words = set(stopwords.words('english'))

# 2. Dataset create
data = """Text;Emotion
i feel so hopeless;sadness
i am grabbing a minute to post this;joy
i feel like i am being punished;sadness
i am feeling overwhelmed;fear
i feel so lucky;joy
i am feeling grumpy;anger
i feel romantic;love
i feel strange;surprise
the situation is very scary;fear
i am so happy today;joy
i feel depressed and lonely;sadness
this is such an amazing gift;joy
i am furious with this service;anger
i feel a bit nervous;fear
i love spending time with you;love
i was shocked by the news;surprise"""

df = pd.read_csv(pd.compat.pickle_compat.io.StringIO(data), sep=';')
df = pd.concat([df] * 100, ignore_index=True) # to increase the data size

# 3. Step-by-Step Cleaning Function
def clean_text(text):
    # Step A: remove no. and speacial characters
    text = re.sub('[^a-zA-Z]', ' ', text).lower()
    # Step B: Tokenization (Words mein todna)
    words = text.split()
    # Step C: Stopwords and Stemming (e.g., feeling -> feel)
    clean_words = [ps.stem(w) for w in words if w not in stop_words]
    return " ".join(clean_words)

# Cleaning Apply
df['Clean_Text'] = df['Text'].apply(clean_text)
print("--- Preprocessing Done! ---")
print(df[['Text', 'Clean_Text']].head()) # Iska screenshot paper mein 'Table 1' banega

# 4. TF-IDF Feature Extraction
tfidf = TfidfVectorizer(max_features=1000)
X_tfidf = tfidf.fit_transform(df['Clean_Text']).toarray()
X = np.reshape(X_tfidf, (X_tfidf.shape[0], 1, X_tfidf.shape[1])) # 3D for BiLSTM

# Labels Encode
le = LabelEncoder()
y = le.fit_transform(df['Emotion'])

# 5. BiLSTM Model Architecture
model = Sequential([
    Bidirectional(LSTM(64, activation='tanh'), input_shape=(1, X.shape[2])),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dense(len(le.classes_), activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# 6. Final Training for Presentation
print("\n--- BiLSTM Training shuru ho rahi hai ---")
history = model.fit(X, y, epochs=10, batch_size=16, validation_split=0.2)

# Results
acc = history.history['accuracy'][-1]
print(f"\nFinal Accuracy: {acc*100:.2f}%")
```

```python
import matplotlib.pyplot as plt

# Accuracy graph
plt.plot(history.history['accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.show()
```
